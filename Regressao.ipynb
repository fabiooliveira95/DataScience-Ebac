```python
import numpy as np
import patsy
import statsmodels.api as sm
```

```python
# %%
# Verificar variável qualitativa, convertendo-as para dummies
# Definindo a fórmula para o modelo
formula = (
    "np.log(renda) ~ C(sexo, Treatment) + C(posse_de_veiculo, Treatment) + "
    "C(posse_de_imovel, Treatment) + C(tipo_renda, Treatment) + "
    "C(educacao, Treatment) + C(estado_civil, Treatment) + "
    "C(tipo_residencia, Treatment) + idade + tempo_emprego + "
    "qt_pessoas_residencia + qtd_filhos"
)

# Preparando os dados com o Patsy
y, X = patsy.dmatrices(formula, df, return_type='dataframe')

# Adicionando uma constante (intercepto) no modelo
X = sm.add_constant(X)

# Ajustando o modelo de regressão OLS
model = sm.OLS(y, X)
result = model.fit()

# Exibindo o resumo do modelo
print(result.summary())

# Avaliando os coeficientes do modelo
print("\nCoeficientes do modelo:")
print(result.params)
```

```python
# Criação de uma função para remover iterativamente as variáveis menos significantes
def iterative_variable_removal(formula, data, significance_level=0.05):
    """
    Ajusta um modelo de regressão removendo iterativamente variáveis não significativas.
    :param formula: String com a fórmula inicial do modelo
    :param data: DataFrame com os dados
    :param significance_level: Nível de significância a ser considerado (default = 0.05)
    :return: Modelo final ajustado e fórmula final
    """
    y, X = dmatrices(formula, data=data, return_type='dataframe')
    model = sm.OLS(y, X).fit()  # Ajusta o modelo inicial
    print("Modelo Inicial:")
    print(model.summary())

    while True:
        # Identifica a variável com maior p-valor
        p_values = model.pvalues
        highest_pvalue_var = p_values.idxmax()
        highest_pvalue = p_values[highest_pvalue_var]

        # Checa se o maior p-valor é maior que o nível de significância
        if highest_pvalue > significance_level:
            print(f"\nRemovendo variável '{highest_pvalue_var}' com p-valor {highest_pvalue:.4f}...")
            # Remove a variável da fórmula
            formula = formula.replace(f" + {highest_pvalue_var}", "").replace(f"{highest_pvalue_var} + ", "")
            # Reajusta o modelo com a nova fórmula
            y, X = dmatrices(formula, data=data, return_type='dataframe')
            model = sm.OLS(y, X).fit()
        else:
            # Sai do loop quando todas as variáveis são significantes
            break

    print("\nModelo Final:")
    print(model.summary())
    return model, formula


# 1. Executa a função de remoção iterativa para ajustar o modelo
final_model, final_formula = iterative_variable_removal(formula, df)

# 2. Comparação dos indicadores entre o modelo inicial e o final
print("\nComparação entre os Modelos:")
print(f"- R² Ajustado (Inicial): {model.rsquared_adj:.4f}, (Final): {final_model.rsquared_adj:.4f}")
print(f"- AIC (Inicial): {model.aic:.2f}, (Final): {final_model.aic:.2f}")
print(f"- BIC (Inicial): {model.bic:.2f}, (Final): {final_model.bic:.2f}")

# Opcional: Visualizar as variáveis removidas e a fórmula final
print("\nFórmula Final do Modelo:")
print(final_formula)
```
