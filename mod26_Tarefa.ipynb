```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
#Plotando
plt.scatter(x = df.p1, y=df.p2);
```

```python
# Calculando média e evolução
df['média'] = df.mean(axis=1)
df['evolução'] = df.p2-df.p1
df.style.format(precision=1, decimal=',')
```

## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
# Calculando soma e diferença
df['soma'] = df.p1 + df.p2
df['diferença'] = df.p2 - df.p1
```

```python
# Plotando o gráfico
from matplotlib import pyplot as plt

plt.figure(figsize=(8, 6)), plt.scatter(df['soma'], df['diferença'], alpha=0.7), plt.title('Relação entre Soma e Diferença de p1 e p2'), plt.xlabel('Soma (p1 + p2)'), plt.ylabel('Diferença (p2 - p1)'), plt.grid(True), plt.show()
```
