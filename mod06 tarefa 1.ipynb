```python
df = pd.read_csv('b1abb33c4f.csv')
df.head(5)

# Calcular o APGAR5 médio por número de CONSULTAS
avg_apgar5_by_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()

# Criar o gráfico de barras
avg_apgar5_by_consultas.plot(kind='bar', color='skyblue')

# Adicionar título e rótulos aos eixos
plt.title('APGAR5 Médio por Número de CONSULTAS')
plt.xlabel('Número de CONSULTAS')
plt.ylabel('APGAR5 Médio')

# Mostrar o gráfico
plt.xticks(rotation=0)  # Mantém os rótulos do eixo x horizontais
plt.show()
```

#### 4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
df = pd.DataFrame(df)

# Criar a figura e os eixos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Gráfico de frequência de SEXO
df['SEXO'].value_counts().plot(kind='bar', ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title('Frequência de SEXO')

# Gráfico de frequência de GESTACAO
df['GESTACAO'].value_counts().plot(kind='bar', ax=axs[0, 1], color='lightgreen')
axs[0, 1].set_title('Frequência de GESTACAO')

# Gráfico de frequência de ESTCIVMAE
df['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axs[1, 0], color='salmon')
axs[1, 0].set_title('Frequência de ESTCIVMAE')

# Gráfico de frequência de CONSULTAS (podemos assumir que é uma contagem ou categórico)
df['CONSULTAS'].value_counts().plot(kind='bar', ax=axs[1, 1], color='gold')
axs[1, 1].set_title('Frequência de CONSULTAS')

# Ajustar layout
plt.tight_layout()
plt.show()
```
