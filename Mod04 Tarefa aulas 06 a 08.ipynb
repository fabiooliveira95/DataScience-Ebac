2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
def consulta_dados(data: List[Dict[str, str]], uf: str, tipo_interessado: str) -> pd.DataFrame:
    """
    Filters the dataset based on the specified state and interested type.

    Args:
    data (List[Dict[str, str]]): The dataset to filter, where each entry is a dictionary.
    uf (str): The state to filter by (e.g., "rio_de_janeiro").
    tipo_interessado (str): The type of interest to filter by (e.g., "Município").

    Returns:
    pd.DataFrame: A DataFrame containing the filtered data.
    """
    # Filter the dataset based on the specified state and type of interest
    dados_filtrados = [entry for entry in data if entry["rio_de_janeiro"] == uf and entry["Tipo"] == tipo_interessado]

    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(dados_filtrados)

    return df
````
